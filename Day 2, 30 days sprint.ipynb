```python
# Personal Data
age_years = 25
weight_kg = 60
height_m = 1.65

# Sleep hours for each day
mon = 5
tue = 7
wed = 6
thurs = 5
fri = 6
sat = 7
sun = 7

# Recommended sleep
recommended_sleep_per_day = 8

# Monthly money
income = 150000
expenses = 90000
```

```python
# convert age to days
age_days = age_years * 365

# convert days to hours
age_hours = age_days * 24

# convert hours to minutes
age_minutes = age_hours * 60

print(age_days, age_hours, age_minutes)
```

```python
# Calculate BMI using weight / height ** 2
bmi = weight_kg / (height_m ** 2)
print ("BMI:" , round(bmi,2))
```

```python
# total sleep for the week

total_sleep = mon + tue + wed + thurs + fri + sat + sun

# Recommeneded weekly sleep
recommended_weekly_sleep = recommended_sleep_per_day * 7

# Difference
sleep_difference = total_sleep - recommended_weekly_sleep

print("total sleep:", total_sleep)
print("recommended:", recommended_weekly_sleep)
print("difference:", sleep_difference)
```

```python
# Budget calculations

savings = income - expenses
# calculate savings percentage
savings_percentage = (savings / income) * 100

# calculate expense percentage
expense_percentage = (expenses / income) * 100
print("Income:",  income)
print("Expenses:", expenses)
print("savings:", savings)
print("savings %:", round(savings_percentage, 2))
print("expense %:", round(expense_percentage, 2))
```
